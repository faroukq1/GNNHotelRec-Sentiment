```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/labr_cleaned_sentiment_camelbert.csv")
print("Dataset shape:", df.shape)
df.head(2)
```

```python
print("Columns after cleanup:")
print(df.columns)
```

```python
df = df[[
'rating',
'review_id', 
'user_id',
'book_id',
'review_text',
'review_text_clean',
'text_length',
'sentiment_label'
]]
print(df.columns)
```

```python
def rating_to_sentiment(r):
    if r in [4, 5]:
        return 1
    elif r == 3:
        return 0
    elif r in [1, 2]:
        return -1
    else:
        return np.nan
df["ground_truth"] = df["rating"].apply(rating_to_sentiment)
```
